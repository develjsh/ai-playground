# 수치형 피처 상관관계 히트맵이란?
목적: 데이터셋에 있는 **수치형 변수 간의 관계(연관성)**를 한눈에 보기 위해 사용합니다.

기본 원리:
상관계수(Correlation Coefficient)는 두 변수 간의 선형 관계를 나타내는 값입니다.
일반적으로 Pearson 상관계수를 사용하며, 값의 범위는 -1 ~ 1입니다.
+1에 가까울수록: 강한 양의 상관관계 (한 값이 증가하면 다른 값도 증가)
-1에 가까울수록: 강한 음의 상관관계 (한 값이 증가하면 다른 값은 감소)
0 근처: 거의 상관관계 없음

**히트맵(Heatmap)**은 이 상관계수를 색상으로 표현하여, 직관적으로 어떤 변수들이 서로 밀접하게 연관되어 있는지 파악하게 해줍니다.


## 분석 시 해석 포인트
높은 상관관계 피처
예: radius_mean와 perimeter_mean가 0.99라면 거의 같은 정보를 주므로, 둘 중 하나를 제거해도 모델 성능이 크게 변하지 않을 수 있음.

음의 상관관계
예: smoothness_worst와 compactness_mean이 -0.75라면 한 피처가 커질수록 다른 피처는 작아짐.

피처 선택(Feature Selection)
상관계수가 매우 높은 피처는 다중공선성(Multicollinearity) 문제를 만들 수 있으니, 회귀 계열 모델에서는 제거 고려.

PCA(차원 축소)
상관관계가 높은 피처가 많다면 PCA로 차원을 줄여도 정보 손실이 적음.

## method='pearson'
method='pearson'은 **피어슨 상관계수(Pearson correlation coefficient)**를 계산하겠다는 의미입니다.

### 피어슨 상관계수란?
두 수치형 변수 사이의 선형(직선) 관계를 측정하는 지표
값의 범위: -1 ~ 1
+1 → 완벽한 양의 상관관계 (한 변수가 증가하면 다른 변수도 같은 비율로 증가)
-1 → 완벽한 음의 상관관계 (한 변수가 증가하면 다른 변수는 같은 비율로 감소)
0 → 선형 상관관계 없음 (하지만 비선형 관계일 수는 있음)



```python
# 수치형 변수만 선택
numeric_features = df.select_dtypes(include=[np.number]).columns

# 상관계수 행렬 계산
corr_matrix = df[numeric_features].corr(method='pearson')

# 히트맵 시각화
plt.figure(figsize=(12,10))
sns.heatmap(
    corr_matrix,        # 상관계수 행렬
    cmap='coolwarm',    # 색상 맵 (파란색: 음의 상관, 빨간색: 양의 상관)
    annot=True,         # 상관계수 숫자 표시
    fmt=".2f",          # 소수점 2자리
    square=True,        # 셀을 정사각형으로
    linewidths=0.5,     # 셀 경계선 두께
    cbar_kws={"shrink": .8}  # 컬러바 크기 조정
)
plt.title('Correlation Heatmap of Numeric Features', fontsize=15)
plt.show()
```

```python
# 수치형 피처 상관관계 히트맵
numeric_features = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12,10))
corr_matrix = df[numeric_features].corr()
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 주요 피처 분포 확인 (예시: 첫 5개 수치형 변수)
for col in numeric_features[:5]:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```
