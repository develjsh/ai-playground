```python
# 라이브러리 불러오기
import pandas as pd              
import numpy as np           
import matplotlib.pyplot as plt  
import seaborn as sns           
%matplotlib inline
```

```python
# sklearn library에서 데이터셋 불러오기
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

## 데이터셋 키 설명

| 키 | 설명 |
| --- | --- |
| **data** | 특징 행렬로, 각 행은 개별 샘플을 나타내며 각 열은 해당 샘플의 특징을 나타냄 |
| **target** | 레이블(타깃) 배열로, 각 샘플에 대한 목푯값(클래스 레이블)이 들어 있음 |
| **feature_names** | 특징들에 대한 이름(컬럼 이름)이 포함된 배열. (이 키가 없을 수도 있음) |
| **DESCR** | 데이터 셋에 대한 자세한 설명이 들어 있는 문자열. 데이터 셋의 출처, 설명, 특징, 레이블 등의 정보 포함 |
| **filename** | 데이터 파일의 경로나 URL이 포함된 문자열 |



```python
print(cancer['feature_names'])
```

```python
# target 확인
print(cancer['target_names'])
print(cancer['target'])
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
df_cancer.head()
```

# 데이터 시각화

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df_cancer)
```
